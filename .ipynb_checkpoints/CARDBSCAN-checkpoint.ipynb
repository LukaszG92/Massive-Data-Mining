```python
import numpy as np
import dbscan_srr as dbscan  # Il tuo modulo
import hdbscan               # Per il calcolo di DBCV
from sklearn.preprocessing import normalize
import time

# 1. Configurazione e Dati
file_path = "/content/drive/MyDrive/DATA MINING/Progetto/CaR/embeddings_v1_384.npy"
# Parametri scelti (Target Similarity ~75% + Micro-cluster)
EPS = 0.71
MIN_PTS = 4
MAX_RAM_GB = 12
N_THREADS = 2
PROB_ERROR = 0.1

print(f"Caricamento dati da {file_path}...")
X = np.load(file_path)

# Normalizzazione (Essenziale per coerenza semantica)
# Nota: SRRDBSCAN lavora in Euclideo, normalizzando diventa "Cosine-like"
X_norm = normalize(X, norm='l2', axis=1)

# 2. Esecuzione SRR-DBSCAN
print(f"\n--- Avvio SRR-DBSCAN (Eps={EPS}, MinPts={MIN_PTS}) ---")
srr = dbscan.SRR()  # Nome classe corretto confermato

start_time = time.time()
labels = srr.fit_predict(
    X_norm,      # Dati normalizzati
    PROB_ERROR,  # probability of missing a point
    MAX_RAM_GB,  # memory constraint in GB
    N_THREADS,   # number of threads
    EPS,         # eps
    MIN_PTS      # minPts
)
end_time = time.time()

# Conversione in numpy array per sicurezza
labels = np.array(labels)

# 3. Statistiche Base
print("\n--- Risultati Clustering ---")
# Stampa le statistiche interne di SRR (es. tempo LSH, memoria)
print(srr.statistics())

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)
noise_ratio = n_noise / len(labels)

print(f"Tempo esecuzione: {end_time - start_time:.2f} sec")
print(f"Cluster trovati: {n_clusters}")
print(f"Noise Points: {n_noise} ({noise_ratio:.2%})")

"""
# 4. Calcolo DBCV (Density-Based Clustering Validation)
# Scientificamente necessario per validare la densità
if n_clusters > 1:
    print("\n--- Calcolo DBCV Score (può richiedere tempo) ---")
    try:
        # hdbscan.validity_index richiede float64
        # Passiamo i dati normalizzati perché il clustering è stato fatto su quelli
        X_eval = X_norm.astype(np.float64)

        dbcv_score = hdbscan.validity.validity_index(X_eval, labels, metric='euclidean')

        print(f"✅ DBCV Validity Score: {dbcv_score:.4f}")

        if dbcv_score >= 0.2:
            print("   -> Qualità: BUONA (Struttura densa rilevata)")
        elif dbcv_score > 0:
            print("   -> Qualità: DEBOLE MA POSITIVA (Meglio del random)")
        else:
            print("   -> Qualità: PESSIMA (Probabile struttura casuale o noise)")

    except Exception as e:
        print(f"Errore nel calcolo DBCV: {e}")
        print("Suggerimento: Controlla che ci sia abbastanza memoria per hdbscan.")
else:
    print("❌ DBCV saltato: Meno di 2 cluster trovati.")
"""
# 5. Salvataggio Output
np.save("final_srr_labels.npy", labels)
print("\nEtichette salvate in 'final_srr_labels.npy'")
```

```python
# Colab-friendly single-cell script for SRRDBSCAN tuning (no labels, no DBSCAN-vs-exact)
# Requirements:
#   - your embeddings saved as a .npy file (already L2-normalized, ideally mean_center+renorm)
#   - the library importable as: import dbscan_srr as dbscan
#
# What it does:
#   1) loads X from INPUT_NPY
#   2) builds eps candidates from k-distance quantiles (k = minPts)
#   3) runs SRRDBSCAN over grid (delta, L_gb, minPts, eps)
#   4) evaluates each run with intrinsic metrics + stability against eps*(1±jitter)
#   5) saves a JSON report + prints Top-K configs

import os, time, json
import numpy as np

import dbscan_srr as dbscan
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score


# -------------------------
# USER CONFIG (edit here)
# -------------------------
INPUT_NPY = "/content/drive/MyDrive/DATA MINING/Progetto/CaR/all-MiniLM-L6-v2_mean_centered.npy"   # <-- change
OUT_JSON  = "/content/drive/MyDrive/DATA MINING/Progetto/CaR/srr_tuning_report1.json"

# SRR params grid
DELTAS   = [0.1, 0.05]     # probability of missing a point
L_GBS    = [1.0]           # memory constraint in GB (per library API)
THREADS  = 1

# DBSCAN params grid
MINPTS_VALUES = [30, 50, 80, 120]
EPS_QUANTILES = [0.05, 0.10, 0.15, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80]
KDIST_METRIC  = "euclidean"  # SRRDBSCAN implementations are typically Euclidean/E2LSH
MAX_EPS_PER_MINPTS = 12

# Stability check (label-free): compare eps vs eps*(1±jitter)
EPS_JITTER = 0.02  # +/-2%

# Ranking / filtering
NOISE_MAX = 0.80
MAX_CLUSTER_SHARE_MAX = 0.50
MIN_CLUSTERS = 2
TOPK = 10

# If your dataset is very large and kNN is slow, you can set a sample size for k-distance estimation:
KDIST_SAMPLE_SIZE = None  # e.g., 20000; None = use full X


# -------------------------
# Helpers
# -------------------------
def cluster_summary(labels: np.ndarray) -> dict:
    labels = np.asarray(labels)
    n = labels.size
    noise = int(np.sum(labels == -1))
    noise_frac = float(noise / n)

    core = labels[labels != -1]
    if core.size == 0:
        return {
            "n_clusters": 0,
            "noise": noise,
            "noise_frac": noise_frac,
            "max_cluster_share": 0.0,
            "median_cluster_size": 0.0,
            "p90_cluster_size": 0.0,
            "p99_cluster_size": 0.0,
        }

    uniq, counts = np.unique(core, return_counts=True)
    counts = np.sort(counts)
    return {
        "n_clusters": int(len(uniq)),
        "noise": noise,
        "noise_frac": noise_frac,
        "max_cluster_share": float(counts[-1] / n),
        "median_cluster_size": float(np.median(counts)),
        "p90_cluster_size": float(np.quantile(counts, 0.90)),
        "p99_cluster_size": float(np.quantile(counts, 0.99)),
    }


def run_srr(X: np.ndarray, delta: float, L_gb: float, n_threads: int, eps: float, minPts: int):
    srr = dbscan.SRR()
    t0 = time.time()
    labels = srr.fit_predict(X, delta, L_gb, n_threads, eps, minPts)
    dt = time.time() - t0
    stats = srr.statistics()
    return labels, float(dt), stats


def stability_metrics(base_labels: np.ndarray, alt_labels: np.ndarray) -> dict:
    # ARI/NMI here are purely between two SRR runs (stability), NOT vs DBSCAN ground truth.
    return {
        "stability_ari": float(adjusted_rand_score(base_labels, alt_labels)),
        "stability_nmi": float(normalized_mutual_info_score(base_labels, alt_labels)),
    }


def compute_eps_candidates_from_kdist(X: np.ndarray, k: int, quantiles: list, metric: str, max_eps: int):
    nn = NearestNeighbors(n_neighbors=k + 1, metric=metric, algorithm="auto")
    nn.fit(X)
    dists, _ = nn.kneighbors(X, return_distance=True)
    kth = dists[:, -1]
    qvals = np.quantile(kth, quantiles)

    eps_candidates = sorted(set(float(v) for v in qvals))
    if len(eps_candidates) > max_eps:
        idx = np.linspace(0, len(eps_candidates) - 1, max_eps).round().astype(int)
        eps_candidates = [eps_candidates[i] for i in idx]

    return {
        "k": int(k),
        "metric": metric,
        "kth_summary": {
            "mean": float(kth.mean()),
            "std": float(kth.std()),
            "min": float(kth.min()),
            "max": float(kth.max()),
        },
        "quantiles": {f"q{int(q*100):02d}": float(v) for q, v in zip(quantiles, qvals)},
        "eps_candidates": eps_candidates,
    }


def rank_runs(runs: list) -> list:
    scored = []
    for r in runs:
        s = r["summary"]
        if s["n_clusters"] < MIN_CLUSTERS:
            continue
        if s["noise_frac"] > NOISE_MAX:
            continue
        if s["max_cluster_share"] > MAX_CLUSTER_SHARE_MAX:
            continue

        stab = r["stability"]
        stab_nmi = 0.5 * (stab["minus"]["stability_nmi"] + stab["plus"]["stability_nmi"])
        stab_ari = 0.5 * (stab["minus"]["stability_ari"] + stab["plus"]["stability_ari"])

        score = (
            2.0 * stab_nmi +
            1.0 * stab_ari -
            1.0 * s["noise_frac"] -
            1.5 * s["max_cluster_share"] -
            0.01 * r["runtime_s"]
        )
        scored.append((score, stab_nmi, stab_ari, r))

    scored.sort(key=lambda x: x[0], reverse=True)
    return scored


# -------------------------
# Main execution (single cell)
# -------------------------
assert os.path.exists(INPUT_NPY), f"Input file not found: {INPUT_NPY}"

X_full = np.load(INPUT_NPY).astype(np.float32)
n, d = X_full.shape
print(f"Loaded X: shape={X_full.shape}")

# Optional: sample for k-distance estimation only
if KDIST_SAMPLE_SIZE is not None and KDIST_SAMPLE_SIZE < n:
    rng = np.random.default_rng(0)
    idx = rng.choice(n, size=KDIST_SAMPLE_SIZE, replace=False)
    X_kdist = X_full[idx]
    print(f"Using k-distance sample: {X_kdist.shape}")
else:
    X_kdist = X_full

kdist_info = {}
eps_grid = {}

print("\n[Step 1] Building eps candidates via k-distance quantiles...")
for minPts in MINPTS_VALUES:
    info = compute_eps_candidates_from_kdist(
        X_kdist, k=minPts, quantiles=EPS_QUANTILES, metric=KDIST_METRIC, max_eps=MAX_EPS_PER_MINPTS
    )
    kdist_info[str(minPts)] = info
    eps_grid[minPts] = info["eps_candidates"]
    print(f"  minPts={minPts:4d}  eps_candidates={len(eps_grid[minPts])}  kth_mean={info['kth_summary']['mean']:.4f}")

runs = []
print("\n[Step 2] Running SRRDBSCAN grid + stability checks...")

for delta in DELTAS:
    for L_gb in L_GBS:
        for minPts in MINPTS_VALUES:
            for eps in eps_grid[minPts]:
                cfg = {"delta": float(delta), "L_gb": float(L_gb), "threads": int(THREADS),
                       "eps": float(eps), "minPts": int(minPts)}

                labels, dt, stats = run_srr(X_full, delta, L_gb, THREADS, eps, minPts)
                summ = cluster_summary(labels)

                eps_minus = eps * (1.0 - EPS_JITTER)
                eps_plus  = eps * (1.0 + EPS_JITTER)

                labels_m, dt_m, _ = run_srr(X_full, delta, L_gb, THREADS, eps_minus, minPts)
                labels_p, dt_p, _ = run_srr(X_full, delta, L_gb, THREADS, eps_plus, minPts)

                run_row = {
                    **cfg,
                    "runtime_s": float(dt),
                    "summary": summ,
                    "statistics": stats,  # library-specific; stored as string if not JSON-serializable
                    "stability": {
                        "eps_minus": float(eps_minus),
                        "eps_plus": float(eps_plus),
                        "runtime_s_minus": float(dt_m),
                        "runtime_s_plus": float(dt_p),
                        "minus": stability_metrics(labels, labels_m),
                        "plus": stability_metrics(labels, labels_p),
                    }
                }
                runs.append(run_row)

                print(f"  done: δ={delta} L={L_gb} minPts={minPts} eps={eps:.6f} "
                      f"clusters={summ['n_clusters']} noise={summ['noise_frac']:.3f} "
                      f"max_share={summ['max_cluster_share']:.3f} time={dt:.2f}s")

report = {
    "meta": {
        "created_at_unix": int(time.time()),
        "input_npy": INPUT_NPY,
        "shape": {"n": int(n), "d": int(d)},
        "kdist_metric": KDIST_METRIC,
        "deltas": DELTAS,
        "L_gbs": L_GBS,
        "threads": THREADS,
        "minPts_values": MINPTS_VALUES,
        "eps_quantiles": EPS_QUANTILES,
        "eps_jitter": EPS_JITTER,
        "max_eps_per_minpts": MAX_EPS_PER_MINPTS,
        "kdist_sample_size": KDIST_SAMPLE_SIZE,
    },
    "kdist": kdist_info,
    "runs": runs,
    "notes": [
        "No ground-truth labels used.",
        "No ARI/NMI vs exact DBSCAN computed.",
        "ARI/NMI are used only as stability measures between SRR runs at eps*(1±jitter).",
        "eps candidates are derived from k-distance quantiles (k=minPts).",
    ]
}

with open(OUT_JSON, "w", encoding="utf-8") as f:
    json.dump(report, f, indent=2, default=str)

print(f"\nSaved report to: {OUT_JSON}")

ranked = rank_runs(runs)
print("\nTop candidates (label-free ranking):")
for i, (score, stab_nmi, stab_ari, r) in enumerate(ranked[:TOPK], start=1):
    s = r["summary"]
    print(f"{i:02d}) score={score:.4f} "
          f"δ={r['delta']} L={r['L_gb']} minPts={r['minPts']} eps={r['eps']:.6f} "
          f"clusters={s['n_clusters']} noise={s['noise_frac']:.3f} max_share={s['max_cluster_share']:.3f} "
          f"stabNMI={stab_nmi:.3f} stabARI={stab_ari:.3f} time={r['runtime_s']:.2f}s")
```

```python
import numpy as np
import pandas as pd
import time
import math

import dbscan_srr as dbscan  # il tuo modulo SRRDBSCAN
from sklearn.preprocessing import normalize
from sklearn.neighbors import NearestNeighbors

SEED = 42
rng = np.random.default_rng(SEED)

print("Imports OK. Seed =", SEED)
```

```python
# 1) Configurazione e Dati
file_path = "/content/drive/MyDrive/DATA MINING/Progetto/CaR/all-MiniLM-L6-v2_mean_centered.npy"

print(f"Caricamento dati da {file_path}...")
X = np.load(file_path)
print("Shape X:", X.shape, "dtype:", X.dtype)

# Normalizzazione L2 (Euclideo ~ Cosine-like)
X_norm = normalize(X, norm='l2', axis=1)

# Sanity check: norme ~ 1
norms = np.linalg.norm(X_norm, axis=1)
print("Norms: min/med/max =", norms.min(), np.median(norms), norms.max())
```

```python
def basic_cluster_stats(labels: np.ndarray):
    labels = np.asarray(labels)
    n = len(labels)
    noise = int(np.sum(labels == -1))
    noise_ratio = noise / n

    core_labels = labels[labels != -1]
    if core_labels.size == 0:
        return {
            "n_clusters": 0,
            "noise": noise,
            "noise_ratio": float(noise_ratio),
            "largest_cluster_frac": 0.0,
            "median_cluster_size": 0.0
        }

    unique, counts = np.unique(core_labels, return_counts=True)
    n_clusters = int(len(unique))
    largest_cluster_frac = float(counts.max() / n)
    median_cluster_size = float(np.median(counts))

    return {
        "n_clusters": n_clusters,
        "noise": noise,
        "noise_ratio": float(noise_ratio),
        "largest_cluster_frac": largest_cluster_frac,
        "median_cluster_size": median_cluster_size
    }


def heuristic_score(res, target_noise=(0.5, 0.15), max_largest_cluster_frac=0.10):
    """
    Score più alto = meglio.

    - preferisci noise_ratio in [0.15, 0.55]
    - evita un cluster dominante (largest_cluster_frac > 0.25)
    - evita n_clusters <= 1
    - penalizza lievemente runtime
    """
    nr = float(res["noise_ratio"])
    lc = float(res["largest_cluster_frac"])
    k  = int(res["n_clusters"])
    t  = float(res["time_sec"])

    # noise term
    if nr < target_noise[0]:
        noise_term = - (target_noise[0] - nr) * 2.0
    elif nr > target_noise[1]:
        noise_term = - (nr - target_noise[1]) * 2.0
    else:
        noise_term = +0.5

    # dominance term
    if lc <= max_largest_cluster_frac:
        dom_term = 0.0
    else:
        dom_term = - (lc - max_largest_cluster_frac) * 3.0

    # cluster count term
    if k <= 1:
        k_term = -3.0
    else:
        k_term = math.log(k + 1e-9) / 5.0

    # time term (soft)
    time_term = - math.log1p(t) / 10.0

    return noise_term + dom_term + k_term + time_term
```

```python
def estimate_eps_candidates(
    X_norm: np.ndarray,
    min_pts: int,
    sample_size=5000,
    quantiles=(0.90, 0.925, 0.95, 0.975)
):
    """
    Stima una lista di ε candidati dalla distribuzione delle distanze
    al min_pts-esimo vicino (k-distance heuristic) su un campione.
    """
    n = X_norm.shape[0]
    m = min(sample_size, n)
    idx = rng.choice(n, size=m, replace=False)
    Xs = X_norm[idx]

    # KNN su campione
    nn = NearestNeighbors(n_neighbors=min_pts + 1, metric="euclidean", algorithm="auto")
    nn.fit(Xs)
    dists, _ = nn.kneighbors(Xs, return_distance=True)

    # dists[:,0] = 0 (self); kth neighbor è colonna min_pts
    kth = dists[:, min_pts]

    eps_list = [float(np.quantile(kth, q)) for q in quantiles]
    eps_list = sorted(set([round(e, 6) for e in eps_list]))

    info = {
        "kth_dist_min": float(kth.min()),
        "kth_dist_med": float(np.median(kth)),
        "kth_dist_max": float(kth.max()),
        "sample_size": int(m)
    }
    return eps_list, info
```

```python
def run_srrdbscan_once(
    X_norm: np.ndarray,
    eps: float,
    min_pts: int,
    prob_error: float,
    max_ram_gb: int,
    n_threads: int
):
    srr = dbscan.SRR()
    t0 = time.time()
    labels = srr.fit_predict(X_norm, prob_error, max_ram_gb, n_threads, eps, min_pts)
    t1 = time.time()

    labels = np.asarray(labels)
    stats = basic_cluster_stats(labels)

    res = {
        "eps": float(eps),
        "min_pts": int(min_pts),
        "prob_error": float(prob_error),
        "time_sec": float(t1 - t0),
        **stats,
        "srr_statistics": str(srr.statistics())
    }
    res["heur_score"] = float(heuristic_score(res))

    return res, labels
```

```python
# Parametri "di sistema" (come i tuoi)
MAX_RAM_GB = 12
N_THREADS = 2

# Spazi di ricerca (puoi restringere/allargare)
min_pts_list = [10, 20]           # micro-cluster -> stabilità
prob_error_list = [0.1, 0.05]     # più piccolo => più robusto (spesso più lento)

# Genera eps candidates data-driven per ciascun minPts e fai union
eps_candidates = set()

for mp in min_pts_list:
    eps_list_mp, info = estimate_eps_candidates(
        X_norm,
        min_pts=mp,
        sample_size=5000,
        quantiles=(0.90, 0.95)
    )
    print(f"minPts={mp} eps candidates={eps_list_mp} | kth-dist stats={info}")
    eps_candidates.update(eps_list_mp)

eps_list = sorted(eps_candidates)
print(f"\nFinal eps_list (union) [{len(eps_list)}]:", eps_list)

# opzionale: restringi un intorno (esempio)
#eps_list = [e for e in eps_list if 0.50 <= e <= 0.80]
#print(f"\nFiltered eps list [{len(eps_list)}]:", eps_list)
```

```python
results = []
label_cache = {}

total = len(eps_list) * len(min_pts_list) * len(prob_error_list)
done = 0

for min_pts in min_pts_list:
    for eps in eps_list:
        for pe in prob_error_list:
            done += 1
            print(f"[{done}/{total}] eps={eps} minPts={min_pts} prob_error={pe}")

            res, labels = run_srrdbscan_once(
                X_norm=X_norm,
                eps=eps,
                min_pts=min_pts,
                prob_error=pe,
                max_ram_gb=MAX_RAM_GB,
                n_threads=N_THREADS
            )

            results.append(res)
            label_cache[(eps, min_pts, pe)] = labels

df = pd.DataFrame(results).sort_values("heur_score", ascending=False).reset_index(drop=True)

print("\nTop 20 configs by heuristic score:")
display(df.head(20))

df.to_csv("srrdbscan_tuning_results.csv", index=False)
print("Saved: srrdbscan_tuning_results.csv")
```

```python
import numpy as np

def basic_report(X, name="X"):
    X = np.asarray(X)
    n, d = X.shape
    norms = np.linalg.norm(X, axis=1)

    print(f"== {name} ==")
    print(f"shape: {X.shape}")
    print(f"norms: mean={norms.mean():.6f} std={norms.std():.6f} "
          f"min={norms.min():.6f} max={norms.max():.6f}")

    # componente media e sua norma: misura grezza di anisotropia
    mu = X.mean(axis=0)
    mu_norm = np.linalg.norm(mu)
    print(f"mean vector norm ||mu||: {mu_norm:.6f}")

    # statistica per coordinate (utile per capire scale e eventuali dimensioni morte)
    coord_mean = X.mean(axis=0)
    coord_std = X.std(axis=0)
    print(f"per-dim std: mean={coord_std.mean():.6f} median={np.median(coord_std):.6f} "
          f"min={coord_std.min():.6f} max={coord_std.max():.6f}")

    return {"n": n, "d": d, "norms": norms, "mu_norm": mu_norm, "coord_std": coord_std}

def pairwise_stats_sample(X, n_pairs=200_000, seed=0):
    rng = np.random.default_rng(seed)
    n = X.shape[0]
    i = rng.integers(0, n, size=n_pairs)
    j = rng.integers(0, n, size=n_pairs)
    # evita i=j
    mask = i != j
    i, j = i[mask], j[mask]

    cos = np.sum(X[i] * X[j], axis=1)  # per unit vectors: dot = cosine
    dist = np.sqrt(np.maximum(0.0, 2.0 - 2.0*cos))

    def q(a): return np.quantile(a, [0.01, 0.05, 0.5, 0.95, 0.99])

    out = {
        "cos_mean": float(cos.mean()),
        "cos_std": float(cos.std()),
        "cos_q": q(cos),
        "dist_mean": float(dist.mean()),
        "dist_std": float(dist.std()),
        "dist_q": q(dist),
    }
    return out

from sklearn.neighbors import NearestNeighbors

def knn_distance_profile(X, k=30, metric="cosine"):
    # k+1 perché il primo vicino è il punto stesso (distanza 0)
    nn = NearestNeighbors(n_neighbors=k+1, metric=metric, algorithm="auto")
    nn.fit(X)
    dists, idx = nn.kneighbors(X, return_distance=True)

    # scarta self-neighbor
    dists = dists[:, 1:]
    idx = idx[:, 1:]
    kth = dists[:, -1]
    return dists, idx, kth

def summarize_array(a, name="a"):
    qs = np.quantile(a, [0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
    print(f"{name}: mean={a.mean():.6f} std={a.std():.6f} "
          f"q01={qs[0]:.6f} q05={qs[1]:.6f} q25={qs[2]:.6f} "
          f"q50={qs[3]:.6f} q75={qs[4]:.6f} q95={qs[5]:.6f} q99={qs[6]:.6f}")
```
